```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.info()
```

```python
color = tips.sex.map({'Female':'red','Male':'green'})
```

```python
color
```

```python
tips.sex.value_counts().plot(kind='bar')
plt.show()
```

```python
sns.countplot(x='sex',data=tips,palette='BuPu')
plt.show()
```

```python
plt.figure(figsize=(10,5),facecolor='#99ff99')
plt.scatter(tips.total_bill,tips.tip,color=color)
plt.text(5,7,'Female',color='red',fontsize=15)
plt.text(5,6,'Male',color='green',fontsize=15)
plt.xlabel('total_bill')
plt.ylabel('tips')
plt.title('tips and total_bill')
plt.show()
```

```python
plt.figure(figsize=(10,5),facecolor='indianred')
sns.scatterplot(x=tips.total_bill,y=tips.tip,hue='sex',data=tips)
plt.show()
```

```python
plt.figure(figsize=(10,5),facecolor='indianred')
sns.scatterplot(x=tips.total_bill,y=tips.tip,hue='sex',size='smoker',data=tips,palette='Set1')
plt.show()
```

```python
plt.figure(figsize=(10,5),facecolor='#800080')
sns.countplot(x='sex',hue='smoker',data=tips,palette='afmhot')
plt.show()
```

```python
plt.figure(figsize=(10,5),facecolor='#800080')
sns.countplot(x='sex',hue='smoker',data=tips,palette='afmhot',dodge=False,alpha=0.6)
plt.show()
```

```python
tips.head()
```

```python
plt.figure(figsize=(10,5),facecolor='hotpink')
sns.barplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
tips.head()
```

```python
plt.figure(figsize=(10,5),facecolor='seagreen')
sns.distplot(tips['total_bill'],bins=30,hist_kws={'color':'g','edgecolor':'k'},
            kde_kws={'color':'k','label':'kde','linewidth':3},rug=True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.jointplot(x='total_bill',y='tip',data=tips,color='red',kind='scatter')
plt.show()
```

```python
sns.regplot(x='total_bill',y='tip',data=tips,color='red')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.swarmplot(x='sex',y='tip',hue='smoker',data=tips)
plt.show()
```

```python
sns.violinplot(x='sex',y='total_bill',data=tips)
plt.show()
```

```python
sns.boxplot(x='sex',y='total_bill',hue='smoker',data=tips,palette='Dark2')
plt.show()
```
