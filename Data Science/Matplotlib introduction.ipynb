```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.figure(figsize=(10,5),facecolor='#ff9999')
plt.plot(x,y)
plt.title('plot of x and y')
plt.xlabel('number of x values')
plt.ylabel('number of x square')
plt.show()
```

```python
a = np.arange(1,6)
b = a**3
```

```python
plt.figure(figsize=(10,5))
plt.plot([1,2,3,4,5],[1,4,9,16,25],'go',a,b,'r^')
#plt.plot(a,b,'r^')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot([1,2,3,4,5],[1,4,9,16,25],'g',marker='*',markersize=10,markerfacecolor='y')
plt.plot(a,b,'r',marker='v',markersize=10,markerfacecolor='y')
plt.title('plot of squares')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.style.use('default')
plt.plot([1,2,3,4,5],[1,4,9,16,25],'g',marker='*',markersize=10,markerfacecolor='y')
plt.plot(a,b,'r',marker='v',markersize=10,markerfacecolor='y')
plt.title('plot of squares')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(linestyle='--',linewidth=0.5)
plt.show()
```

```python
plt.subplot(1,2,1) #row,column,acces of graph
plt.plot(x,y,'r:')


plt.subplot(1,2,2)
plt.plot(y,x,'g:')
plt.show()
```

```python
fig = plt.figure(figsize=(10,5),facecolor='#99ff99')

ax1 = fig.add_axes([0,0,0.8,0.8])  #left,bottom,width,height
ax2 = fig.add_axes([0.15,0.35,0.3,0.3])

ax1.plot(x,y,'m:')
ax1.set_title('Large plot')
ax1.set_xlabel('x-axis')
ax1.set_ylabel('y-axis')

ax2.plot(y,x,'y:')
ax2.set_title('Small plot')
ax2.set_xlabel('x-axis')
ax2.set_ylabel('y-axis')

plt.show()
```
