```python
with open('table.txt') as f:
    data = f.read()
    f.close()
```

```python
data
```

```python
data = data.split('\n')
data
```

```python
dat = list(map(lambda string : string.split(),data))
dat
```

```python
for s,t,u,v in dat:
    print(s,t,u,v,sep='\t')
```

```python
mat = []
sci = []
for s,t,u,v in dat:
    mat.append(u[:])
    sci.append(v[:])
    
print(mat,sci,sep='\n')
```

```python
mat.remove('math')
sci.remove('science')

```

```python
ma = []
sc = []
for var in mat:
    ma.append(int(var))
    
for tar in sci:
    sc.append(int(tar))
```

```python
print(ma,sc,sep='\n')
```
